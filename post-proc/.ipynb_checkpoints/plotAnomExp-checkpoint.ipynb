```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import matplotlib.animation as animation
import glob
import xarray as xr
import cv2
```

```python
def plot_waveIce(fileCICE, fileWW3, xG, yG, listV, repOut, dateStr):
    
    #orig_map=plt.cm.get_cmap('gist_earth')
    orig_map=plt.cm.get_cmap('Spectral')
    reversed_map = orig_map.reversed()
        
    print("File wave : ", fileWW3)
    print("File ice : ", fileCICE)
    
    dsW3=xr.open_dataset(fileWW3)
    hs_2d=np.nan_to_num(dsW3[['hs']]['hs'].values.squeeze())
    
    dsCI=xr.open_dataset(fileCICE)
    
    #cbar_min=0
    #cbar_max=300
    #cbarlabels = np.linspace(np.floor(cbar_min), np.ceil(cbar_max), num=5, endpoint=True)
    
    #Read wave field
#     hs_2d=np.genfromtxt(file_hs, dtype=None, skip_header=1)
#     hs_rst=np.zeros((len(xG), len(yG)))
#     hs_2d=cleanGrdFiles_ww3(hs_2d, hs_rst, defV, convF)
    
    X,Y = np.meshgrid(xG,yG)
    fig, ax = plt.subplots(len(listV),1,figsize=[8,4*len(listV)])
    i=0
    for v in listV:
        var_2d=np.nan_to_num(dsCI[[v]][v].values.squeeze())  
        ax[i].set_xlim([2.5, 240])
        ax[i].set_ylim([2.5, 240])
        ax[i].tick_params(labelsize=14)
#         ax[i].set_yticks(np.arange(0.5, 3, 0.5))
        ax[i].set_xlabel('x [km]', size=16)
        ax[i].set_ylabel('y [km]', size=16)
        cont=ax[i].contour(X,Y, hs_2d, colors='black', levels=[0.01, 0.02, 0.04, 0.06, 0.1, 0.14, 0.2])
        color=ax[i].contourf(X,Y,var_2d, 50, cmap=reversed_map)
        ax[i].clabel(cont, fontsize= 12)
        cb=fig.colorbar(color, ax=ax[i])
        cb.ax.tick_params(labelsize=14)
        
        if i == 0:
            cb.set_label('Ice Concentration', size=16)
    
        if i == 1:
            cb.set_label('Ice Thickess [m]', size=16)
        if i == 2:
            cb.set_label('Mean Floe Diameter [m]', size=16)
        i=i+1
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.97)
    fig.savefig(repOut+"/waveIce_"+dateStr, bbbox_to_anchor='tight', dpi=500)
    #ig.savefig("/aos/home/bward/wim/post_proc/case10/test", bbbox_to_anchor='tight', dpi=500)
    print("------------"+repOut+"/waveIce"+dateStr+" as been plotted---------------------")

def plotAnom(fileCICE_1, fileWW3_1,fileCICE_2, fileWW3_2, xG, yG, listV, repOut, dateStr, exp1, exp2):
    orig_map=plt.cm.get_cmap('seismic')
    reversed_map = orig_map.reversed()
    
    print("Anom (wave) between  : ", fileWW3_1, fileWW3_2 )
    print("Anom (ice) between : ", fileCICE_1, fileCICE_2)
    
    dsW3_1=xr.open_dataset(fileWW3_1)
    dsW3_2=xr.open_dataset(fileWW3_2)
    
    hs_2d_1=np.nan_to_num(dsW3_1[['hs']]['hs'].values.squeeze())
    hs_2d_2=np.nan_to_num(dsW3_2[['hs']]['hs'].values.squeeze())
    
    anom_hs=hs_2d_2-hs_2d_1
    
    dsCI_1=xr.open_dataset(fileCICE_1)
    dsCI_2=xr.open_dataset(fileCICE_2)
  
    #cbar_min=0
    #cbar_max=300
    #cbarlabels = np.linspace(np.floor(cbar_min), np.ceil(cbar_max), num=5, endpoint=True)
    
    #Read wave field
#     hs_2d=np.genfromtxt(file_hs, dtype=None, skip_header=1)
#     hs_rst=np.zeros((len(xG), len(yG)))
#     hs_2d=cleanGrdFiles_ww3(hs_2d, hs_rst, defV, convF)
    
    X,Y = np.meshgrid(xG,yG)
    fig, ax = plt.subplots(len(listV),1,figsize=[8,4*len(listV)])
    i=0
    for v in listV:
        var_2d_1=np.nan_to_num(dsCI_1[[v]][v].values.squeeze())
        
        if v == 'aice':
            vW3='ice'
        elif v == 'hi':
            vW3='ic1'
        elif v == 'fsdrad':
            vW3='ic5'
            
        var_2d_2=np.nan_to_num(dsCI_2[[vW3]][vW3].values.squeeze())
        
        anom_var=var_2d_2-var_2d_1
        
        ax[i].set_xlim([2.5, 240])
        ax[i].set_ylim([2.5, 240])
        ax[i].tick_params(labelsize=14)
#         ax[i].set_yticks(np.arange(0.5, 3, 0.5))
        ax[i].set_xlabel('x [km]', size=16)
        ax[i].set_ylabel('y [km]', size=16)
        cont=ax[i].contour(X,Y, anom_hs, colors='black', levels=[-0.1, -0.06, -0.04, -0.02, -0.01, 0.01, 0.02, 0.04, 0.06, 0.1])
        color=ax[i].contourf(X,Y,anom_var, 50, cmap=orig_map,  vmin = -10, vmax =10)
        ax[i].clabel(cont, fontsize= 12)
        cb=fig.colorbar(color, ax=ax[i])
        cb.ax.tick_params(labelsize=14)
        
        if i == 0:
            cb.set_label('Ice Concentration', size=16)
    
        if i == 1:
            cb.set_label('Ice Thickess [m]', size=16)
        if i == 2:
            cb.set_label('Mean Floe Diameter [m]', size=16)
        i=i+1
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.97)
    fig.savefig(repOut+"/anom_"+exp1+"_"+exp2+"_"+dateStr, bbbox_to_anchor='tight', dpi=500)
    print("------------"+repOut+"/anom_"+exp1+"_"+exp2+"_"+dateStr+" as been plotted---------------------")

def createListTs(ts, nts, sec_init):
    list_ts=[]
    
    for i in range(1,nts+1):
        list_ts.append(sec_init+i*ts)
            
    return list_ts

def mergeImage(plot1, plot2, repOut, dateStr, exp1, exp2):
    #Read the two images

    image1=cv2.imread(plot1)
    image2=cv2.imread(plot2)
    fig = plt.figure(figsize=(16, 12))
    fig.add_subplot(1, 2, 1)
    plt.imshow(image1)
    plt.axis('off')
    plt.title("Coupled")
    
    fig.add_subplot(1, 2, 2)
    plt.imshow(image2)
    plt.axis('off')
    plt.title("Uncoupled")
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.97)
    fig.savefig(repOut+"/panel_waveIce_"+dateStr+"_"+exp1+"_"+exp2, bbbox_to_anchor='tight', dpi=500)
    print("------------"+repOut+"/panel_waveIce_"+dateStr+"_"+exp1+"_"+exp2+".png as been plotted---------------------")
```
